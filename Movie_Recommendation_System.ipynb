```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
print(ratings.head())
print(movies.head())
```

```python
merged_data=pd.merge(ratings,movies,on='movieId')
```

```python
merged_data
```

```python
user_movie_matrix=merged_data.pivot_table(index='userId',columns='title',values='rating')
user_movie_matrix.fillna(0, inplace=True)
user_movie_matrix.head()
```

```python
user_similarity=cosine_similarity(user_movie_matrix)
```

```python
user_similarity_df=pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)
user_similarity_df.head()
```

```python
def recommend_movies(user_id, user_movie_matrix, user_similarity_df, num_recommendations=5):
    # Get the similarity scores for the target user
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)

    # Take the most similar users, excluding the user itself (at index 0)
    similar_users = similar_users.iloc[1:]

    # Calculate weighted ratings based on similar users
    weighted_ratings = pd.Series(dtype=np.float64)

    for similar_user, similarity in similar_users.items():
        rated_movies = user_movie_matrix.loc[similar_user]
        weighted_ratings = weighted_ratings.add(rated_movies * similarity, fill_value=0)

    # Get the user's current ratings to exclude already rated movies
    user_ratings = user_movie_matrix.loc[user_id]

    # Get a list of recommendations by finding movies the user hasn't rated (NaN)
    recommendations = weighted_ratings[user_ratings == 0]

    # Sort the recommendations
    recommendations = recommendations.sort_values(ascending=False).head(num_recommendations)

    # Return the titles of the top recommended movies
    return recommendations.index

# Test the recommendation function for userId 1
recommended_movies = recommend_movies(2, user_movie_matrix, user_similarity_df)
print("Recommended Movies:", recommended_movies)
```
